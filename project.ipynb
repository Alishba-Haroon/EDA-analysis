# Data Analysis and Visualizations


```python
df.columns
```

```python
# Countplot Graph of Order Priority
ax = sns.countplot(x = 'Order Priority',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Countplot of Discount
ax = sns.countplot(data = df, x ='Discount', hue = 'Discount')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
# Barplot of Sales or Profit
new_Profit = df.groupby(['Sales'], as_index=False)['Profit'].sum().sort_values(by='Profit',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = new_Profit, x = 'Sales',y='Profit')
```

```python
# Barplot of Unit Price and Shipping Cost
new_Profit = df.groupby(['Unit Price'], as_index=False)['Shipping Cost'].sum().sort_values(by='Shipping Cost',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = new_Profit, x = 'Unit Price',y='Shipping Cost')
```

```python
# Line Plot
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='Quantity ordered new', y='Product Base Margin')
plt.title('Line Plot')
plt.xlabel('Quantity ordered new')
plt.ylabel('Product Base Margin')
plt.show()
```

```python
# Assuming you have a DataFrame df with columns 'Region'

# Calculate the count of each unique value in the 'Region' column
region_counts = df['Region'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Region Distribution')
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Order Date')
plt.title('Histogram of Order Dates')
plt.xlabel('Order Date')
plt.ylabel('Frequency')
plt.show()


```

```python
# Plotting the jointplot
sns.jointplot(data=df, x='Region', y='Product Container', kind='hist')
plt.show()
```
